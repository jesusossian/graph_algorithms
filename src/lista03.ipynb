```python
import networkx as nx
import matplotlib.pyplot as plt
```

### lista 03

```python
def criar_arvore_binaria_completa(n):

    G = nx.DiGraph()
    
    # add vertices
    for i in range(n):
        G.add_node(i)
        
    # add arestas
    for i in range(n):
        filho_esquerda = 2 * i + 1
        filho_direita = 2 * i + 2
        
        if filho_esquerda < n:
            G.add_edge(i, filho_esquerda)
        if filho_direita < n:
            G.add_edge(i, filho_direita)
            
    return G
```

```python
pos = {
    0: (0, 10),
    1: (-5, 7.5),
    2: (5, 7.5),
    3: (-7.5, 5),
    4: (-2.5, 5),
    5: (2.5, 5),
    6: (7.5, 5)
}
```

```python
n = 7
G = criar_arvore_binaria_completa(n)

#pos = nx.spring_layout(G)
#pos = nx.kamada_kawai_layout(G)
#nx.draw(G, pos, with_labels=True)
#plt.show()
```

```python
plt.figure(figsize=(10, 8))
nx.draw(G, pos, with_labels=True, arrows=True, node_size=5000, font_size=30)
plt.savefig("binary_tree.png")
plt.axis('off')
plt.show()
```
