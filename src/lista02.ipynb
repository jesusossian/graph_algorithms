```python
import networkx as nx
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
```

```python
nodes = [1,2,3,4,5]
edges = [(1,3),(1,4),(1,5),(2,3),(2,4),(2,5),(3,5),(4,5)]
```

```python
G.add_edges_from(edges)
```

```python
#nx.draw(G, with_labels=True)
pos = nx.spring_layout(G)
plt.figure(figsize=(14, 10)) 
nx.draw_networkx(G, pos, with_labels=True, node_color='skyblue', node_size=1000, font_size=10, font_weight='bold')
plt.show()
```

A excentricidade corresponde a maior distncia entre 2 ns conectados.
$$
ecc(v) = \max_{u \in V \setminus{v}} d(v,u) \text{ para } v \in V
$$

O dimetro de um grafo  a sua maior excentricidade.
$$
diametro(G) = \max_{v \in V} ecc(v)
$$

O raio de um grafo  sua menor excentricidade.
$$
raio(G) = \min_{v \in V} ecc(v)
$$

A periferia de um grafo, seria o conjunto de todos os ns cujo a excentricidade  igual ao dimetro.
$$
periferia(G) = \{ v \in V \; | \: ecc(v) = diametro(G) \}
$$

Centro de um grafo  o conjunto de ns que sua excentricidade  igual ao raio.
$$
centro(G) = \{v \; | \; ecc(v) = raio(G) \}
$$

```python
eccentricity_G = nx.eccentricity(G)
print("Eccentricity of all nodes:", eccentricity_G)
```

```python
diametre_G = nx.diameter(G)
print("Diameter of G", diametre_G)
print([k for k,v in nx.eccentricity(G).items() if v == nx.diameter(G)])
```

```python
radius_G = nx.radius(G)
print("Radius of G:", radius_G)
print([k for k,v in nx.eccentricity(G).items() if v == nx.radius(G)])
```

```python
nx.center(G)
```

```python
G.clear()
```
