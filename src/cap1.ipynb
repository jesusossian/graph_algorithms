### Algoritmo 1.1: 
- Inversão de uma sequência
- Dados: sequência s[1],...,s[n]

```python
def inversao_sequencia(s):
    n = len(s)-1 # primeiro elemento eh desconsiderado
    for j in range(1,n//2+1):
        s[j], s[n-j+1] = s[n-j+1], s[j]
```

```python
lista = [None,7,3,1,5,9,10]
```

```python
inversao_sequencia(lista)
```

```python
print(lista)
```

### Algoritmo 1.3: 
- Ordenação de uma sequência
- Dados: sequência s[1],...,s[n]

```python
def ordenacao_sequencia(s):	
	n = len(s)-1
	for j in range(1,n):
		k = j
		while k >= 1 and s[k] > s[k+1]:
			s[k], s[k+1] = s[k+1], s[k]
			k = k-1
```

```python
lista = [None,7,3,1,5,9]
```

```python
ordenacao_sequencia(lista)
```

```python
print(lista)
```
